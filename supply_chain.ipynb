```python
import pandas as pd
import numpy as np
print("Libraries Imported")
```

```python
file_path = 'SCMS_Delivery_History_Dataset.csv'
shipping_data = pd.read_csv(file_path)

shipping_data.head()
```

```python
print("\n Duplicate rows: ", shipping_data.duplicated().sum(),
           " \n \n Columns with Null Values:  \n", shipping_data.isnull().sum().sort_values(ascending=False),
           "\n\n Dataset Info: \n", shipping_data.info())
```

```python
# Convert date columns to datetime
date_columns = [
    'Scheduled Delivery Date',
    'Delivered to Client Date', 
    'Delivery Recorded Date', 
    'PQ First Sent to Client Date', 
    'PO Sent to Vendor Date'
    ]
              
for col in date_columns:
    shipping_data[col] = shipping_data[col].replace('Date Not Specified', np.nan)
    shipping_data[col] = pd.to_datetime(shipping_data[col], errors='coerce', format='%d-%b-%y')

# Convert 'Weight (Kilograms)' and 'Freight Cost (USD)' to numeric, coercing errors
shipping_data['Weight (Kilograms)'] = pd.to_numeric(shipping_data['Weight (Kilograms)'], errors='coerce').fillna(0)
shipping_data['Freight Cost (USD)'] = pd.to_numeric(shipping_data['Freight Cost (USD)'], errors='coerce').fillna(0)

# Fill missing 'Line Item Insurance (USD)' with 0 and 'Shipment Mode' with 'Unknown'
shipping_data['Line Item Insurance (USD)'] = shipping_data['Line Item Insurance (USD)'].fillna(0)
shipping_data['Shipment Mode'] = shipping_data['Shipment Mode'].fillna('Shipment Mode Not Specified') 

print("\nDataFrame Info after transformations:")
print(shipping_data.info())
print("\nFirst 5 rows with new calculated columns:")
print(shipping_data[['PO Sent to Vendor Date', 'Scheduled Delivery Date', 'Delivered to Client Date']].head())
```

```python
# Clean column names: lowercase, replace spaces with underscores
shipping_data.columns = shipping_data.columns.str.strip().str.lower().str.replace(" ", "_")
```
